__init__ 의 역할?
클래스로부터 객체를 만들 때 자동으로 실행되는 초기 설정 함수

```python
class Student:                      #Student라는 설계도를 만든 것
    university = "Korea University"

    def __init__(self, name):
        self.name = name
    def set_age(self, age):
        self.age = age
    def set_major(self, major):
        self.major = major
```

```python
anna = Student('anna')
```

```python
anna.university
```

```python
anna.name
```

```python
anna.set_major('physics')
```

The MasterStudent class inherits the attributes and methods from the previous Student class.

```python
class MasterStudent(Student):
    internship = 'mandatory, from March to June'
```

```python
james = MasterStudent('james')
```

```python
james.internship
```

What if you want MasterStudent to have its own __init__ method? In that case, you can still call the parent class's constructor by using super().__init__(...) inside the new __init__ so that all the initialization from Student (like setting name) is preserved while adding any extra attributes specific to MasterStudent.

MasterStudent가 자기만의 함수를 갖게 하는 방법?

```python
class MasterStudent(Student):
    internship = 'mandatory, from March to June'

    def __init__(self, name, gender):
        super().__init__(name)
        self.gender = gender
```

```python
jiwon = MasterStudent('jiwon', 'female')
```

클래스와 객체지향 프로그래밍 덕분에,
우리가 마주치는 다양한 객체들(예: Model 클래스, Data 클래스 등)에 대응하는
서로 다른 클래스들로 코드를 체계적으로 구성할 수 있다.
각 클래스는 고유한 메서드와 속성을 가진다.

그리고 클래스 상속을 사용하면
하나의 기본 클래스(base class)를 바탕으로 변형된 클래스들을 만들 수 있으며,
코드를 재사용할 수 있다.
예를 들어, 기본 Model 클래스에서 더 발전된 CNNModel 클래스와 같은 것을 만들 수 있다.
