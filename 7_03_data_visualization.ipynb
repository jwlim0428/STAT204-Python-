### 7.3.1. Matplotlib

- A basic tool for visualization in Python ([https://matplotlib.org](https://matplotlib.org/stable/index.html))

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

There are two coding styles:

- **Explicit:** Less intuitive but offers greater flexibility for customization


```python
x = np.linspace(0, 2, 100)  # Sample data.

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

- **Implicit:** More intuitive but less convenient for customization


```python
plt.figure(figsize=(5, 2.7), layout='constrained') # optional
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

### 7.3.2. Explicit Approach

- Figure object: An (empty) object used for creating a plot, on which all the subplots (axes) are placed.

```python
#fig = plt.figure()
fig = plt.figure(figsize = (5,5), layout = 'constrained')
```

```python
fig
```

- Figure is just an empty container, so you need to create at least one subplot
- Use `add_subplot` to create subplots one at a time

```python
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

# Output (empty objects)
fig
```

- At this point, `ax1` to `ax3` are still empty objects — now let’s draw the plots

```python
# Univariate trend
ax1.plot(np.random.standard_normal(50).cumsum(), color="black", linestyle="dashed")

# Histogram
ax2.hist(np.random.standard_normal(100), bins=20, color="black", alpha=0.3)

# Satter Plot
ax3.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))

# Output
fig
```

- To create subplots all at once, use `subplots`

```python
fig, axes = plt.subplots(2, 2, sharex = False, sharey = False)
axes[0,0].plot(np.random.standard_normal(50).cumsum(),
               color="black",
               linestyle="dashed")

axes[0,1].hist(np.random.standard_normal(100),
               bins=20,
               color="black",
               alpha=0.3)

axes[1,0].scatter(np.arange(30),
                  np.arange(30) + 3 * np.random.standard_normal(30))
```

- Main options of `matplotlib.pyplot.subplots`


| argument | description                                                 |
| -------- | ----------------------------------------------------------- |
| nrows  | Integer – Number of rows in the subplot grid                |
| ncols  | Integer – Number of columns in the subplot grid             |
| sharex | Boolean – Whether all subplots should share the same x-axis |
| sharey | Boolean – Whether all subplots should share the same y-axis |

**4. Line Plot with Uncertainty**: `fill_between(x, y1, y2)`
- [https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.fill_between.html#matplotlib.axes.Axes.fill_between](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.fill_between.html#matplotlib.axes.Axes.fill_between)

```python
# make data
np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x)) # upper values
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x)) # lower values
x, y1, y2
```

```python
# Explicitly
fig, ax = plt.subplots()
ax.plot(x, (y1 + y2)/2, linewidth=2) # Line
```

```python
# Explicitly
fig, ax = plt.subplots()
ax.plot(x, (y1 + y2)/2, linewidth=2) # Line
ax.fill_between(x, y1, y2, alpha=.6) # Uncertainty
```

```python
# Implicitly
plt.plot(x, (y1 + y2)/2, linewidth=2) # Line
plt.fill_between(x, y1, y2, alpha=.5) # Uncertainty
```

### 7.3.5. Plots on a mesh grid

**1. Gridded Surface Plot**: `pcolormesh(x, y, z)`

```python
# make data with uneven sampling in x
x = [-3, -2, -1.6, -1.2, -.8, -.5, -.2, .1, .3, .5, .8, 1.1, 1.5, 1.9, 2.3, 3]
y = np.linspace(-3, 3, 128)
X, Y = np.meshgrid(x, y)
X, Y
```

```python
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
```

```python
# plot
fig, ax = plt.subplots()
ax.pcolormesh(X, Y, Z)
```

```python
# or
plt.pcolormesh(X, Y, Z)
```

**2. Contour Plot**: `contour(x, y, z, levels)` or `contourf(x, y, z, levels)`

```python
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# plot
fig, ax = plt.subplots(1, 2, figsize=(10, 5), layout='constrained')
ax[0].contour(X, Y, Z, levels=levels)  # lines
ax[1].contourf(X, Y, Z, levels=levels) # colors
```

### 7.3.6. Adjusting plots

**1. Subplot Sizing and Adjustment**

- By default, appropriate spacing and padding are provided
- To modify these, use the `subplot_adjust` method ([https://matplotlib.org/stable/api/\_as\_gen/matplotlib.pyplot.subplots\_adjust.html](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html))


```python
fig, axes = plt.subplots(2, 2, sharex = True, sharey = True)
for i in range(2):
   for j in range(2):
      axes[i, j].hist(np.random.standard_normal(500),
                      bins = 50,
                      color = "black",
                      alpha = 0.5)
fig.subplots_adjust(wspace = 0, hspace = 0)
```

    - The simplest way to add a legend is to provide the `label` argument to each plot

```python
fig, ax = plt.subplots()
ax.plot(np.random.randn(1000).cumsum(), color="black", label="one")
ax.plot(np.random.randn(1000).cumsum(), color="black", linestyle="dashed", label="two")
ax.plot(np.random.randn(1000).cumsum(), color="black", linestyle="dotted", label="the other")

ax.legend()
```
