###5.1.6 Variable number of parameters

Special forms of parameters:

*args: any number of positional arguments packed into a tuple.

**kwargs: any number of keyword arguments packed into a dictionary.

```python
def variable_args(*args, **kwargs):
    print('args is', args)
    print('kwargs is', kwargs)
```

```python
variable_args(1, 2, 3, a="one", b="two", c="three")
```

Syntax:

lambda arguments : expression (only one line allowed)

```python
x = lambda a : a + 10    #lambda는 한 줄로
```

```python
x(5)
```

```python
x = lambda a, b : a * b
```

```python
x(5, 6)
```

```python
nums = [1, 2, 3, 4, 5]
squared = map(lambda x: x**2, nums)  #map함수는 리스트의 각 원소에 같은 함수를 하나씩 적용
```

1. Write a function stats(a, b, c) that takes three numbers and returns three values (their sum, their average, and their maximum) as a tuple.

```python
def stats(a, b, c):                #stat이라는 함수
    total = a + b + c              #총합
    average = total / 3            #평균
    maximum = max(a, b, c)         #최댓값
    return total, average, maximum
```

```python
x, y, z = stats(3, 5, 10)
```

```python
print(x, y, z)
```

2. Write a function my_sum that takes any number of numeric arguments and returns their sum.

```python
def my_sum(*args):
    total = 0
    for i in args:
        total += i
    return total
```

3. Write a small program using at least two functions:

A function is_prime(n) that returns True if n is a prime number and False otherwise.

A function list_primes(limit) that uses is_prime to return a list of all prime numbers less than or equal to limit.

Show sample output for list_primes(20).

```python
def is_prime(n):
    if n < 2:
        return False
    else:
        for i in range(2, int(n / 2) + 1):    # consider integers 2, ..., int(n / 2)
            if (n % i == 0):
                return False
        return True
```

```python
is_prime(4)
```

```python
def list_primes(limit):
    primes = []
    for i in range(1, limit + 1):
        if is_prime(i):
            primes.append(i)

    return primes
```

```python
list_primes(10)
```
